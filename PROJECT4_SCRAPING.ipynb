# MULTI-PAGE

```python
for index in range (1,23):
    page = requests.get(f'https://www.distributorwardah.com/kategori-produk/page/{index}')
    soup = BeautifulSoup(page.content,'html.parser')
```

```python
for index in range (1,23):
    print(f'https://www.distributorwardah.com/kategori-produk/page/{index}')
```

```python
data_produk=[]
data_HargaAwal=[]
data_HargaDiskon=[]

for index in range (1,23):
    page = requests.get(f'https://www.distributorwardah.com/kategori-produk/page/{index}')
    soup = BeautifulSoup(page.content,'html.parser')
    
    produk = soup.find_all("h2", class_="card-title titleproduk")
    for each_produk in produk:
        data_produk.append(each_produk.text)
    
    HargaAwal = soup.find_all("p", class_="coret black-text mb0")
    for each_HargaAwal in HargaAwal:
        data_HargaAwal.append(each_HargaAwal.text)
    
    HargaDiskon = soup.find_all("p", class_="teal-text mb0")
    for each_HargaDiskon in HargaDiskon:
        data_HargaDiskon.append(each_HargaDiskon.text)
        
#data diskon tidak digunakan karena pada dataframe "All arrays must be of the same length"
```

```python
data={
    'Nama Produk':data_produk,
    'Harga Awal':data_HargaAwal,
    'Harga Diskon':data_HargaDiskon
}
```

```python
df=pd.DataFrame(data)
```
